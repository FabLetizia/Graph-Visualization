```python
import json
```

```python
# Leggi il file
with open('graph.txt', 'r') as file:
    lines = file.readlines()

# Dizionari per mantenere la corrispondenza tra vecchi e nuovi ID
node_mapping = {}
current_id = 1

# Liste per i dati dei nodi e dei collegamenti
nodes = []
links = []
communities = set()

# Parsing delle linee
for line in lines:
    parts = line.strip().split('\t')
    if len(parts) < 3:
        continue

    old_from = int(parts[0])
    old_to = int(parts[1])
    community = int(parts[2])
    communities.add(community)

    # Associa nuovi ID se necessario
    if old_from not in node_mapping:
        node_mapping[old_from] = current_id
        nodes.append({"key": current_id, "text": f"Node {current_id}"})
        current_id += 1
    if old_to not in node_mapping:
        node_mapping[old_to] = current_id
        nodes.append({"key": current_id, "text": f"Node {current_id}"})
        current_id += 1

    # Aggiungi il collegamento con i nuovi ID e la comunità
    links.append({"from": node_mapping[old_from], "to": node_mapping[old_to], "community": community})

# Crea il dizionario finale da passare a GoJS
graph_data = {
    "nodeDataArray": nodes,
    "linkDataArray": links,
    "numCommunities": len(communities)  # Numero di comunità
}

# Salva i dati in formato JSON
with open('graph.json', 'w') as json_file:
    json.dump(graph_data, json_file, indent=4)

print("Conversione completata. Dati salvati in grafo.json.")
```
